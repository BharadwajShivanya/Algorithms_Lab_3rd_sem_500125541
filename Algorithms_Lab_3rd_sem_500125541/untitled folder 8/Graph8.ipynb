```python
import numpy as np
import matplotlib.pyplot as plt
import time

def subset_sum(nums, target):
    n = len(nums)
    # Create a 2D array to store results of subproblems
    dp = np.zeros((n + 1, target + 1), dtype=bool)

    # If the target is 0, then the answer is true (empty set)
    for i in range(n + 1):
        dp[i][0] = True

    # Fill the subset sum table
    for i in range(1, n + 1):
        for j in range(1, target + 1):
            if nums[i - 1] <= j:
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - nums[i - 1]]
            else:
                dp[i][j] = dp[i - 1][j]

    return dp[n][target]

def measure_time(num_elements, target):
    nums = np.random.randint(1, 20, num_elements)  # Random integers between 1 and 20
    start_time = time.time()
    result = subset_sum(nums, target)
    end_time = time.time()
    return end_time - start_time

# Parameters for plotting
target = 30
num_sizes = 20
times = []

# Measure execution time for varying sizes of input
for i in range(1, num_sizes + 1):
    time_taken = measure_time(i * 10, target)  # Increase size by 10 each time
    times.append(time_taken)
    print(f"Size: {i * 10}, Time taken: {time_taken:.6f} seconds")

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(range(10, (num_sizes + 1) * 10, 10), times, marker='o')
plt.title('Subset Sum Algorithm Performance')
plt.xlabel('Input Size (Number of Elements)')
plt.ylabel('Time Taken (seconds)')
plt.grid()
plt.xticks(range(10, (num_sizes + 1) * 10, 10))
plt.show()
```
